```python
def dict(letter):
    dict = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    return dict.index(letter)
def undict(number):
    dict = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    return dict[number]
```

```python
def cov(str):
    output = []
    for i in str:
        output.append(dict(i))
    return output
```

```python
def en(lst):
    output = []
    for i in range(0,len(lst)):
        if i == 0:
            output.append(lst[1])
        elif i == len(lst)-1:
            output.append(lst[-2])
        else:
            output.append((lst[i-1]+lst[i+1])%26)
    return output
            
        
        
   
```

```python
def de(lst):
    output = []
    for i in lst:
        output.append(undict(i))
        
    return output
```

```python
#GENERATE DATASET
X = []
Y = []
for i in range(1,5000):
    data = random.sample(range(26), 4)
    Y.append(data)
    X.append(en(data))
X = np.array(X)
Y = np.array(Y)    
```

```python
model = Sequential()
model.add(Dense(4,input_dim=X.shape[1],activation= 'relu'))
for i in range(1,3):
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(0.2))
model.add(Dense(Y.shape[1]))
model.compile(loss= "MSE",optimizer='adam',metrics=['accuracy'])


startime= time.time()
history = model.fit(X,Y,epochs= 1000,batch_size = 50,verbose = 2, validation_split= 0.2,)
endtime = time.time()
print("time:")
print(endtime-startime)

```
